```python
from bs4 import BeautifulSoup, NavigableString

with open("legacy.aonprd.com/indices/feats.html", 'r') as file:
    feats_html = BeautifulSoup(file.read())
feats_html
```

```python
feats_table = feats_html.find('table')
feats_table
```

```python
feats_rows = feats_table.find_all('tr')
feats_rows
```

```python
feats_cells = [row.find_all('td') for row in feats_rows]
feats_cells
```

```python
columns = ['URI', 'Name', 'Prerequisites', 'Short Description']
feats = []
row: list[BeautifulSoup]
for row in feats_cells:
    if len(row) >= 3:
        anchor = row[0].find('a')
        name = anchor.get_text() if anchor else row[0].get_text().strip()
        sup = row[0].find('sup')
        if sup and sup.get_text().strip() == 'M':
            name += ' - Mythic'
        feats.append({
            'URI': anchor.attrs['href'] if anchor else None,
            'Name': name,
            'Prerequisites': row[1].get_text().strip(),
            'Short Description': row[2].get_text().strip()
        })
feats
```

```python
import pandas as pd

feats_df = pd.DataFrame(feats)
feats_df
```

```python
feat_uri = feats_df['URI'][0]
feat_uri_split = feat_uri.split("#")
feat_page_path = feat_uri_split[0].replace('..', 'legacy.aonprd.com')
feat_id = feat_uri_split[1]

feat_name = ""
feat_text = ""
with open(feat_page_path, 'r', encoding='utf-8') as file:
    feat_page = BeautifulSoup(file.read())
    feat_header = feat_page.find('h2', id=feat_id)
    feat_name = feat_header.get_text().strip()
    feat_paragraphs = []
    next_paragraph = feat_header.next_sibling
    while next_paragraph.name != 'h2':
        feat_paragraphs.append(next_paragraph)
        next_paragraph = next_paragraph.next_sibling
    feat_text = "\n".join([f.get_text().strip() for f in feat_paragraphs]).strip()
print(feat_name)
feat_text
```

```python
from bs4 import NavigableString

typos = {
    'alter-binary-mindscape': 'alter-bindary-minscape',
    'aquatic-adaptation': 'aquatic-adaption',
    'attuned-to-the-wild': 'attuned-to-the-wind',
    'guardian-of-the-wild': 'guardian-of-the-wind',
    'battlefield-hearler': 'battlefield-healer-(story)',
    'critcal-cure': 'critical-cure',
    'critical-versatility': 'critical-vesatility-(combat)',
    "destroyers-blessing": "destroyer's-blessing-(combat)",
    "distruptive-mythic": "disruptive-mythic",
    "improved duck and cover": "improved-duck-and-cover",
    "exiles-path": "exile's-path",
    'adaptive-fortune': 'adaptive-fortunte',
    'grasping-tail': 'grasphing-tail',
    "horn-riders-charge": "horn-rider's-charge",
    "lifes-blood": "life's-blood",
    "lucky-halfling": "lucky-halflin",
    "mothers-gift": "mother's-gift",
    "mythic-companion-mythic": "mythic-companion",
    "snipers-lantern": "sniper's-lantern",
    "technology-adept": "technology-adept-combat",
    "robots-bane": "robot-bane",
    "scavengers-eye": "scavenger's-eye",
    "scavengers-luck": "scavanger-luck",
    "self-sufficient-mythic": """self-sufficient-mythic
""",
    "missle-shield-mythic": "missile-shield-mythic",
    "seige-gunner": "siege-gunner",
    "signatureSkill": "signature-skill",
    "greater skald's vigor": "greater-skalds-vigor",
    "leaping-shot": "leaping-shot-deed",
    "board-ferocity": "boar-ferocity",
    "board-shred": "boar-shred",
    "crane-reposte": "crane-riposte",
    "diviner's-delvin": "diviner's-delving",
    "spider-climber": "spider-step",
    "spirit-of-the-corps": "spirit-of-the-corps-(combat,-teamwork)",
    "spirit's gift": "spirits-gift",

}

url_typos = {
    "../advancedPlayersGuide/advancedFeats.html#blood-of-heroes": "../advancedPlayersGuide/advancedNewRules.html#blood-of-heroes",
    "../advancedPlayersGuide/advancedFeats.html#luck-of-heroes": "../advancedPlayersGuide/advancedNewRules.html#luck-of-heroes",
    'http://paizo.com/unchained/skillsAndOptions/consolidatedSkills/index.html#skill-focus': '../unchained/skillsAndOptions/consolidatedSkills/index.html#skill-focus-unchained',
}

def get_feat_text(feat_uri):
    # print(feat_uri)
    if not feat_uri:
        print('No Feat URI')
        return ''
    if feat_uri in url_typos:
        feat_uri = url_typos[feat_uri]
    feat_uri_split = feat_uri.split("#")
    feat_page_path = feat_uri_split[0]
    if 'paizo.com' in feat_page_path:
        feat_page_path = feat_page_path.replace("http://paizo.com/pathfinder/prd", 'legacy.aonprd.com')
    else: # http://paizo.com/pathfinder/prd/ultimateCombat/ultimateCombatFeats.html#leaping-shot
        feat_page_path = feat_page_path.replace('..', 'legacy.aonprd.com')
    feat_id = feat_uri_split[1]

    feat_text = ""
    with open(feat_page_path, 'r', encoding='latin-1') as file:
        feat_paragraphs = []

        feat_page = BeautifulSoup(file.read())
        feat_header = feat_page.find(id=feat_id)
        if not feat_header: # story feats are weirdly formatted
            feat_header = feat_page.find(id=f"{feat_id}-(story)")
        if not feat_header: # some combat feats are weirdly formatted
            feat_header = feat_page.find(id=f"{feat_id}-(combat)")
        if not feat_header: # some teamwork feats are weirdly formatted
            feat_header = feat_page.find(id=f"{feat_id}-(teamwork)")
        if not feat_header: # typos in the data
            if feat_id in typos:
                feat_header = feat_page.find(id=typos[feat_id])
        #print(f"Header: {feat_header}")
        try:
            following_tags = feat_header.next_siblings
            sibling: BeautifulSoup
            for sibling in following_tags:
                if type(sibling) is NavigableString:
                    feat_paragraphs.append(sibling)
                    continue
                if 'id' in sibling.attrs:
                    break
                feat_paragraphs.append(sibling)

            feat_text = "\n".join([f.get_text().strip() for f in feat_paragraphs]).strip()
        except Exception as e:
            print(e)
            print(f"{feat_page_path} | {feat_id} ----- {feat_header}")
    return feat_text
```

```python
from tqdm import tqdm

long_descriptions = []

for i in tqdm(range(len(feats_df))):
    long_descriptions.append(get_feat_text(feats_df['URI'][i]))
long_descriptions
```

```python
feats_df['long_description'] = long_descriptions
feats_df
```

```python
feats_df.to_csv('feats.csv', index=False)
```
